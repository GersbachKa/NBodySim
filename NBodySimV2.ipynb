```python
sim1 = Simulator(name="Sun-Mercury",notebook=True)
```

```python
mercuryMass = 3.285e23
mercuryRadius = 2.43e6
#above is good!

perihelion = 46.002e9
maxSpeed = 58.98e3

mercuryPosition = [-perihelion,0,0]
mercuryVelocity = [0,-maxSpeed,0]

sim1.addObject(name="Sun",mass=1.989e30,radius=696.34e6,position=[0,0,0],velocity=[0,0,0],color=(239,142,56))
sim1.addObject(name="Mercury",mass=mercuryMass,radius=mercuryRadius,position=mercuryPosition,velocity=mercuryVelocity,color=(219,206,202))
```
